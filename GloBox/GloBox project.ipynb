at first glance we can see that from the people who have converted, the people in group A have generated more per user. But it could be becouse of luck(at random) we need to do statstical test to determine if there is a difference in revenue generated by each group.

```python
# making the necceseray list to make our t test
spent_A = df_r[df_r["group"] == "A"]["spent"]
spent_B = df_r[df_r["group"] == "B"]["spent"]

# importing the test we are going to use
from scipy import stats
t_stat, p_value = stats.ttest_ind(spent_A, spent_B, equal_var=False)  
print(t_stat, p_value)
```
